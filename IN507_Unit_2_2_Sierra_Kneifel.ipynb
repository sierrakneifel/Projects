```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Connect to google drive
from google.colab import drive
drive.mount('/content/drive')

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/BusinessData.csv')
```

```python
# skim() provides visual summary of dataframe
from skimpy import skim
skim(df)
```

```python
# Check for missing values
print('Missing values:', df.isna().sum())
```

```python
# Handle missing values by dropping rows with missing values
df.dropna(inplace=True)
```

```python
# Check for missing values
print('Missing values:', df.isna().sum())
```

```python
# Check for outliers
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
sns.boxplot(df['Age'], ax=ax[0][0])
sns.boxplot(df['Income'], ax=ax[0][1])
sns.boxplot(df['Productivity'], ax=ax[1][0])
sns.boxplot(df['Satisfaction'], ax=ax[1][1])
plt.show()
```

```python
# Handle outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Check for outliers
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
sns.boxplot(df['Age'], ax=ax[0][0])
sns.boxplot(df['Income'], ax=ax[0][1])
sns.boxplot(df['Productivity'], ax=ax[1][0])
sns.boxplot(df['Satisfaction'], ax=ax[1][1])
plt.show()
```

```python
replace_dict = {'PD': 'PhD', 'MMM': 'Master'} #replacing the values in the wrong categories
df['Education Level'].replace(replace_dict, inplace=True)
```
